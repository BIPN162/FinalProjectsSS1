## Abstract

   Two datasets acquired from the Allen Institute were explored and analyzed individually in order to examine if the Dorsolateral Prefrontal Cortex (DLPFC) is representative of the entire Prefrontal Cortex (PFC). A microarray dataset from 6 human donors showcases the genetic expression for genes of interest linked to schizophrenia in specific brain regions, and the Mouse Connectivity dataset can be used to yield projection matrices to many regions in the brain from the two brain regions of interest (DLPFC, PFC). For the microarray dataset, the genetic expression data for the DLPFC, anatomically located in the Middle Frontal Gyrus (MFG), was compared to the other regions in the PFC, the Superior Frontal Gyrus (SFG) and Inferior Frontal Gyrus (IFG), via ANOVA one way tests to determine the similarity between the groups. (Mylius et al., 2013) Alternately, in order to compare human brain regions to mouse brain regions in the Mouse Connectivity dataset, the chosen mouse brain regions of interest are predicated on previous studies to equate human DLPFC and PFC (Carlen, 2017). From these regions, projection matrices were derived by plotting heat maps of normalized projection density data in order to examine how the homologous brain regions connect to other regions of the mouse brain. Visual comparisons and percentage differences between the projection matrices of the DLPFC and PFC homologous regions aid in inferring the feasibility of the PFC being represented by the DLPFC. We conclude that the genetic comparison yielded similarity between the DLPFC and the PFC, while the connectivity exploration yielded no discernable result.

# Background and Prior Work

Schizophrenia is a mental disorder that is highly heritable. There does not exist a definite genetic indicator for schizophrenia, however observations from clinical, neuroimaging, and postmortem studies implicate that the primate dorsal prefrontal cortex as a possible locus for cognitive disturbances in Schizophrenia (Mirnics et al., 2000). Microarray studies have enabled scientists to examine expression levels of schizophrenia patients and thereby allow them to identify schizophrenia-associated genes in the dorsolateral prefrontal cortex (Vawter et al., 2002). In particular, genes HINT1, UBE2N, GRIA2, MAPK14, AURKA, GABRD, EPS8L1 are associated with decreased expression among schizophrenic patients and NDUFA4, ST3GAL6, NAIP, FAS are associated with increased expression among schizophrenia patients (Vawter et al., 2002). This identifies them as schizophrenia related genes in terms of gene expression level. Thus using the Microarray data from the Allen Institute, we are able to investigate differences in the expression levels of our genes of interest between DLPFC and PFC. 

Furthermore, the successful development of future treatments for schizophrenia is also dependent on making anatomical connections between mice and humans. Studies have shown that the the medial prefrontal cortex in mice, which is composed of the infralimbic, prelimbic and anterior cingulate cortices, is homologous to the  primate dorsolateral prefrontal cortex (DFPLC (Mirnics et al., 2000). In fact, the components of the medial prefrontal cortex are associated with a distinct function in schizophrenia. Functional neuroimaging revealed the anterior cingulate cortex (ACC) was mostly hypoactive in schizophrenic patients when performing memory and executive tasks (Adams and David, 2007).  On the contrary, the infralimbic (IL) and prelimbic (PL) cortices are associated with social interaction (Minami et al., 2017). In particular, the IL neurons are activated when rodents during leaving behavior which is suggestive of passive relief from cautionary objects  (Minami et al., 2017). Meanwhile, PL neurons are activated during approaching and contact behavior which is suggestive that they are associated with active actions towards others(Minami et al., 2017).

Currently, there are a few problems which exist with studying schizophrenia in rodents. Firstly, schizophrenia is not a naturally occurring disease in mice, instead, scientists have to induce neurological deficits in mice in order to mimic symptoms of schizophrenia (Jones et al., 2011). There has been success with mouse models that have genetic modifications that result in  behavioral phenotype changes, which mimics ‘positive-like’ symptoms  of schizophrenia however inducing disturbances in social interaction and learning and memory which makes up the negative and cognitive symptoms of schizophrenia remains more difficult to induce, however not impossible (Jones et al., 2011). The second problem scientists have encountered lies within the limits of defining the rodent prefrontal cortex because it varies with the tasks that are performed (Swanson et al., 2018). In our project we defined the prefrontal cortex to include the secondary motor areas, anterior cingulate area, prelimbic area, infralimbic area, orbital area, and agranular insulate area (Carlen, 2017). To further investigate whether the  rodent prefrontal cortex is representative of the DLPFC, we pulled the Mouse Connectivity data from the Allen Brain Institute in order to compare the connectivity between the regions. Studying the connection between the DLPFC and the PFC could influence future Schizophrenia studies.

## Hypothesis


If the DLPFC is a representative structure for the entire PFC, then genetic data for humans in the regions should showcase similarity in their genetic expression, while mouse connectivity data of homologous mouse brain regions should show similar normalized projection density to the same regions across the mouse brain. 

We expect to see the genetic expression of the selected genes in different regions to result in similar expression across the regions due to numerous previous studies that isolate the DLPFC as a region of interest for studying genetic expression differences in schizophrenia patients. In terms of the mouse connectivity dataset, we expect to see differences in the DLPFC and PFC  homologous mouse brain regions in terms of normalized projection density to other regions in the brain, however we expect to see them projecting to the same regions.

## Data Wrangling: Microarray

```python
service = 'http://api.brain-map.org/api/v2/data/query.json?criteria=' 

#create function to return gene expression level data for one donor
def get_one(donor_str, brain_str, gene_str, probe_str):

    #To get the donor ID
    specimen_name = donor_str
    donor_result = requests.get('%smodel::Donor[name$eq\'%s\']' % (service,specimen_name)).json()
    donor_id = donor_result['msg'][0]['id']

    # To get the ID for a brain area
    brain_area = brain_str
    ontology_id = 'Human Brain Atlas'
    structure_result = requests.get('%smodel::Structure,\
        rma::criteria,[name$il\'%s\'],ontology[name$eq\'%s\']' % (service,brain_area,ontology_id)).json()
    structure_id = structure_result['msg'][0]['id']

    # To get the ID for a probe
    gene_acronym = gene_str
    probe_type = probe_str
    product_abbrev = 'HumanMA'

    result = requests.get('%smodel::Probe, \
        rma::criteria,[probe_type$eq\'%s\'],products[abbreviation$eq\'%s\'],gene[acronym$eq\'%s\']' \
                          % (service,probe_type,product_abbrev,gene_acronym)).json()
    probe_id = ''
    for i in range(len(result['msg'])):
        probe_id = probe_id + str(result['msg'][i]['id'])
        if(i != len(result['msg'])-1):
            probe_id = probe_id + ', '
    
    # To pull all this together to get the combination we want
    request_command = '%sservice::human_microarray_expression \
    [probes$eq' + probe_id + '][donors$eq\'%s\'][structures$eq\'%s\']'
    results = requests.get(request_command % (service,donor_id,structure_id)).json()
    
    # Results for all of the probes we tested for our donor
    microarray_df = pd.DataFrame(results['msg']['probes'])
    #print(microarray_df)
    data = microarray_df['expression_level'].values.tolist()
    one_df = []
    #convert data into a list for easier analysis
    for i in range(len(data)):
        for b in range(len(data[i])):
            one_df.append(float(data[i][b]))
    return one_df
```

```python
#create a function that calls get_one function to compile gene expression level from all six donors into one list
def get_all_donor(brainStr, geneStr, probeStr='DNA', is_list=True):
    df1 = get_one(donor_str='H0351.1009', brain_str=brainStr, gene_str=geneStr, probe_str=probeStr)
    df2 = get_one(donor_str='H0351.2001', brain_str=brainStr, gene_str=geneStr, probe_str=probeStr)
    df3 = get_one(donor_str='H0351.2002', brain_str=brainStr, gene_str=geneStr, probe_str=probeStr)
    df4 = get_one(donor_str='H0351.1012', brain_str=brainStr, gene_str=geneStr, probe_str=probeStr)
    df5 = get_one(donor_str='H0351.1015', brain_str=brainStr, gene_str=geneStr, probe_str=probeStr)
    df6 = get_one(donor_str='H0351.1016', brain_str=brainStr, gene_str=geneStr, probe_str=probeStr)
    result_all = df1+df2+df3+df4+df5+df6
    return result_all
```

## Data Analysis & Results: Microarray

We will run one way ANOVA tests in order to compare the mean gene expression across the subregions.

```python
#Running ANOVA test 
hint1_fvalue, hint1_pvalue = stats.f_oneway(hint1_inf_df,hint1_mid_df,hint1_sup_df)
ube2n_fvalue, ube2n_pvalue = stats.f_oneway(ube2n_inf_df,ube2n_mid_df,ube2n_sup_df)
gria2_fvalue, gria2_pvalue = stats.f_oneway(gria2_inf_df,gria2_mid_df,gria2_sup_df)
mapk14_fvalue, mapk14_pvalue = stats.f_oneway(mapk14_inf_df,mapk14_mid_df,mapk14_sup_df)
aurka_fvalue, aurka_pvalue = stats.f_oneway(aurka_inf_df,aurka_mid_df,aurka_sup_df)
gabrd_fvalue, gabrd_pvalue = stats.f_oneway(gabrd_inf_df,gabrd_mid_df,gabrd_sup_df)
eps8l1_fvalue, eps8l1_pvalue = stats.f_oneway(eps8l1_inf_df,eps8l1_mid_df,eps8l1_sup_df)
ndufa4_fvalue, ndufa4_pvalue = stats.f_oneway(ndufa4_inf_df,ndufa4_mid_df,ndufa4_sup_df)
st3gal6_fvalue, st3gal6_pvalue = stats.f_oneway(st3gal6_inf_df,st3gal6_mid_df,st3gal6_sup_df)
naip_fvalue, naip_pvalue = stats.f_oneway(naip_inf_df,naip_mid_df,naip_sup_df)
fas_fvalue, fas_pvalue = stats.f_oneway(fas_inf_df,fas_mid_df,fas_sup_df)
```

```python
#Print out the ANOVA result
ANOVA_df = pd.DataFrame(np.array([[hint1_fvalue, ube2n_fvalue, gria2_fvalue, mapk14_fvalue, aurka_fvalue, gabrd_fvalue, eps8l1_fvalue, ndufa4_fvalue, 
                                  st3gal6_fvalue, naip_fvalue, fas_fvalue], [hint1_pvalue, ube2n_pvalue, gria2_pvalue, mapk14_pvalue, aurka_pvalue, gabrd_pvalue, 
                                  eps8l1_pvalue, ndufa4_pvalue, st3gal6_pvalue, naip_pvalue, fas_pvalue]]), columns=[ 'HINT1', 'UBE2N', 'GRIA2', 'MAPK14', 
                                  'AURKA', 'GABRD', 'EPS8L1', 'NDUFA4', 'ST3GAL6', 'NAIP', 'FAS'], index = ['F-value', 'p-value'])
print(ANOVA_df)
```

The Microarray studies have been used to explore the differential expression of schizophrenia associated genes. From a previous study, we selected a set of genes that showed expression levels that significantly varied from the norm in schizophrenia patients(Vawter et al., 2002).  The genes of interest which showed decreased levels of expression include HINT1, UBE2N, GRIA2, MAPK14, AURKA, GABRD, and EPS8L1 (Vawter et al., 2002). The genes of interest which showed increased levels of expression include NDUFA4, ST3GAL6, NAIP, and FAS (Vawter et al., 2002).
```python
#create a function to return a dataframe of structures the mouse PFC projected to, and its corresponsing mean normalized density value
def mean_median(my_experiment, name_regions):
    pm = mcc.get_projection_matrix(experiment_ids = my_experiment, 
                                   hemisphere_ids = [3],
                                   parameter = 'normalized_projection_volume')
    pm1 = mcc.get_projection_matrix(experiment_ids = my_experiment, 
                                   hemisphere_ids = [3],
                                   parameter = 'volume')
    structure_list = list(mcc.default_structure_ids)
    df1234 = pd.DataFrame(pm['matrix']/pm1['matrix'])
    reg_means=list (df1234.mean(axis=0))
    regmean_df = pd.DataFrame({ 'structure id':structure_list, 'value': reg_means})
    return(regmean_df)

#create a function to visualize the dataframe created above
def plot_hist_raw (my_experiment, name_regions):
    pm = mcc.get_projection_matrix(experiment_ids = my_experiment, 
                                   hemisphere_ids = [3],
                                   parameter = 'normalized_projection_volume')
    pm1 = mcc.get_projection_matrix(experiment_ids = my_experiment, 
                                   hemisphere_ids = [3],
                                   parameter = 'volume')
    structure_list = list(mcc.default_structure_ids)
    df1234 = pd.DataFrame(pm['matrix']/pm1['matrix'])
    reg_means=list (df1234.mean(axis=0))
    regmean_df = pd.DataFrame({ 'structure id':structure_list, 'value': reg_means})
    mean_1= float(sum(reg_means)/len(reg_means))
    median_1= float(pd.DataFrame(reg_means).median())
    print(name_regions+' median: '+str(median_1)+ ' mean: '+ str(mean_1))
    plt.hist(reg_means, bins = 30)
    plt.title('Mean Normalized Projection Density Distribution To Every Brain Structure From '+ name_regions)
    plt.xlabel('Normalized Projection Density')
    plt.ylabel('Count')
    plt.show()

#pull data using the function
aca_raw = mean_median(my_experiment = aca_experiment_ids, name_regions = 'ACA')
pl_raw = mean_median(my_experiment = pl_experiment_ids, name_regions = 'PL')
ai_raw = mean_median(my_experiment = ai_experiment_ids, name_regions ='AI')
ila_raw = mean_median(my_experiment = ila_experiment_ids, name_regions = 'ILA')
orb_raw = mean_median(my_experiment = orb_experiment_ids, name_regions = 'ORB')
mos_raw = mean_median(my_experiment = mos_experiment_ids, name_regions = 'MOs')
```

The cell below has a function to eliminate projected structures with the lowest 50% of the normalized projection density.

```python

def structure_elim(raw_mean):
    post_mean = raw_mean[raw_mean['value']>=raw_mean['value'].median()]
    return(post_mean)

#call the method to eliminate structures, while also store the structures that passed the eliminations as dataframe with
#their ids and mean normalized projection density value
aca_pm = structure_elim(raw_mean=aca_raw)
pl_pm = structure_elim(raw_mean=pl_raw) 
ai_pm = structure_elim(raw_mean=ai_raw)
ila_pm = structure_elim(raw_mean=ila_raw)
orb_pm = structure_elim(raw_mean=orb_raw)
mos_pm = structure_elim(raw_mean=mos_raw)

#Inner merge the datas together to create a dataframe for DLPFC and PFC
final_structures_dl = aca_pm
final_structures_dl = final_structures_dl.merge(pl_pm, on='structure id')
final_structures_dl = final_structures_dl.merge(ila_pm, on='structure id')
#DLPFC includes 93 rows at this point
final_structures_pfc = final_structures_dl
final_structures_pfc = final_structures_pfc.merge(ai_pm, on='structure id')
final_structures_pfc = final_structures_pfc.merge(orb_pm, on='structure id')
final_structures_pfc = final_structures_pfc.merge(mos_pm, on='structure id')
#PFC includes 61 rows, these 61 structures will be the structures we look at in the heatmap
final_structures_dl.columns = ['structure_id', 'ACA_passed_value','PL_passed_value','ILA_passed_value']
final_structures_pfc.columns = ['structure_id', 'ACA_passed_value','PL_passed_value','ILA_passed_value','AI_passed_value','ORB_passed_value','MOs_passed_value']
print(final_structures_dl)
print(final_structures_pfc)
```

```python

#store the list of 61 brain regions abbreviation corresping to the structure ids
structure_id_for_raw_hm = final_structures_pfc['structure_id']

#create a function to graph heatmap of all experiments for the mouse PFC regions with their normalized projection density value towards the 61 brain regions
#It also return column labels that contain 61 brain structures' symbols
def graph_heatmap(ctx_children):
    pm = mcc.get_projection_matrix(experiment_ids = aca_experiment_ids +pl_experiment_ids +ila_experiment_ids + mos_experiment_ids + ai_experiment_ids + orb_experiment_ids, 
                                   projection_structure_ids = ctx_children,
                                   hemisphere_ids= [3], # right hemisphere, ipsilateral
                                   parameter = 'normalized_projection_volume')
    pm2 = mcc.get_projection_matrix(experiment_ids = aca_experiment_ids +pl_experiment_ids +ila_experiment_ids + mos_experiment_ids + ai_experiment_ids + orb_experiment_ids, 
                                   projection_structure_ids = ctx_children,
                                   hemisphere_ids= [3], #Show both hemisphere combined data
                                   parameter = 'volume')
    row_labels = pm['rows'] # these are just experiment ids
    column_labels = [ c['label'] for c in pm['columns'] ] 
    matrix = pm['matrix']/pm2['matrix']
    
    fig, ax = plt.subplots(figsize=(15,15))
    heatmap = ax.pcolor(matrix, cmap=plt.cm.afmhot)
    
    # put the major ticks at the middle of each cell
    ax.set_xticks(np.arange(matrix.shape[1])+0.5, minor=False)
    ax.set_yticks(np.arange(matrix.shape[0])+0.5, minor=False)
    
    ax.set_xlim([0, matrix.shape[1]])
    ax.set_ylim([0, matrix.shape[0]])          
    
    # want a more natural, table-like display
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    
    ax.set_xticklabels(column_labels, minor=False, rotation = 45)
    ax.set_yticklabels(row_labels, minor=False)
    plt.show()
    return(column_labels)
```

## Conclusion & Discussion

Our primary interest in this project was to elaborate on the relationship between the DLPFC and PFC, and see if the PFC is represented by the DLPFC. We investigated this by analyzing the similarities and differences between the regions using two approaches: genetic expression differences in genes of interest, which are predicated on studies exploring Schizophrenia associated gene expression in the regions, as well as connectivity differences of regions homologous to the PFC and DLPFC in the mouse brain. After the two-pronged investigation, we are not able to conclude that the DLPFC is representative of the PFC.

Our results from the Microarray dataset demonstrate the significance of gene expression as it pertains to each region of the prefrontal cortex (Inferior Frontal Gyrus (IFG), Middle Frontal Gyrus (MFG), and Superior Frontal Gyrus (SFG)). In order to ascertain if the DLPFC, which is anatomically located in the MFG, is representative of the other regions in the PFC, the IFG and SFG, one way ANOVA tests were performed on the region subgroups per gene. From the aforementioned statistical tests, we were able to see that ten out of the eleven genes had similar expressions (in terms of means and variances) across the three brain regions (IFG, MFG, and SFG), where the gene AURKA was the only gene that had a p-value less than 0.05, implying significant difference between the MFG, SFG, and IFG. This result indicates that the DLPFC might be a fitting genetic representation for the PFC, a facet of overall representation that must be addressed.  

Conversely, the heat maps constructed from the Mouse Connectivity dataset implied that the primate DLPFC is not represented by the rodent PFC. The normalized projection densities of each analogous brain region projecting to all regions in the brain (as captured by the Mouse Connectivity Dataset) was calculated by dividing the normalized projection volume by the volume. Then we plotted the distribution of the data for each region, in order to assess if the data need to be filtered utilizing the mean or median. From the plots, we see skewed data, and as a result we used the median to filter out the normalized projection density data that is below the median, in order to maintain the data points with high projection density to be displayed on the heatmap. Each filtered projected region, containing only the 50% highest normalized projection density data, was combined to the appropriate dataframe, based on the aforementioned homologous regions(Carlen, 2017). We were then able to plot a heatmap of the raw data (Figure 3) with individual experiments representing each row, and then find the means for the DLPFC and PFC datasets in order to plot a heatmap comparing the DLPFC and PFC’s mean normalized projection density to the selected structures of the brain (Figure 4). Then, the percentage difference between the two regions’ data was calculated for each projected brain structure, where the average percent difference was 14.39%. The variability in those values, along with the visual comparison between the heatmaps indicate that the DLPFC has far more normalized connections than the PFC. Thus, the connectivity dataset showcases indiscernible results to support the hypothesis.  

This project does encounter limitations that should be addressed, and taken into account in future research that is similar in concept. Firstly, many of the published research referenced within the project had defined their results, but not showcased their raw data, resulting in the inability to compare the data of these studies with the data we have procured, and the conclusions we have reached. Moreover, the datasets from which we procured our data was not as expansive as desired. For example, in order to study the connectivity of the DLPFC and PFC, we needed to use the mouse dataset instead of human datasets. The human PFC and the homologous mouse regions are not perfectly representative of each other. Yet, we must warrant that we do have background research indicating the chosen regions as prominent regions of comparison between mice and humans, although not perfectly (Carlen, 2017). Additionally, the projection matrices derived cannot be easily analyzed, because we have no code that could produce any sort of statistical test for the heatmaps, so we utilized percentage differences to compare the data, which is not the favorable approach.

In terms of future directions for DLPFC and PFC comparisons, further research must be conducted on different facets of the brain regions in order to achieve a profound understanding and result, as well as further studies on the topics covered in this project. An initial undertaking that had to be removed from this project was to address the transcriptomics dataset provided by the Allen Brain Institute, which could aid in determining cell types within brain regions, as well as gene expression associated within the cell types. Once an SDK becomes available for this dataset, it would be interesting to compare the cell types of the transcriptomics data set within the regions to the genetic expression within the same region, to observe cell type and gene expression patterns in the DLPFC and PFC. Transcriptomics could be a useful approach in analyzing the regions. Moreover, further studies could look into the electrophysiology of the neurons in the regions, to compare the activity of the regions. This could be connected to the work done for the mouse connectivity, as the connectivity difference could also elaborate on the electrophysiological differences that might be present between the regions. 

Although the etiology of Schizophrenia is not known or understood, our results contribute to the overall understanding of schizophrenia-associated genes and brain structure. In addition, our results may also be applicable in some future treatment options for schizophrenic patients, as mice are the first test subjects before treatment is administered to humans. Thus, defining analogous brain regions in mice and humans is necessary in order to develop drugs which would benefit schizophrenia in humans. Our results yielded that the rodent DLPFC as we defined was not a clear representation of the rodent PFC. Thus, further studies must be conducted to induce all the known schizophrenic symptoms in mice in order to identify the brain regions that are affected. Until the boundaries of the rodent brain regions analogous to the human DLPFC are defined, treatment localized to the DLPFC for schizophrenia and other neurological diseases are long ways from studies.

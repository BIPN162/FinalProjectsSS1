# Background and Prior Work

Memory is a topic which has been studied in different fields, including psychological and biological fields, for a long time, but the mechanism and work processing of it is still not completely clear. The amygdala, hippocampus, and entorhinal cortex in the brain of vertebrates, including humans, are structures that are responsible for emotional processes, learning, and memory (Amunts, et al., 2005).  For this project, we would like to learn more about the amygdala to investigate deep into the mechanism. 

Amygdala is one of two almond-shaped structures in the temporal lobe of the human brain. Moreover, the amygdala is not a single homogeneous structure; instead, it contains various subnuclei based on cytoarchitecture, myeloarchitecture, and chemoarchitecture (Amunts, et al., 2005).The amygdala has numerous afferent and efferent connections with the diencephalon, the mediofrontal, orbitofrontal and lateral prefrontal cortical regions, hypothalamic nuclei, olfactory brain regions, basal ganglia and the brain stem with many reciprocal connections(Amunts, et al., 2005). The basolateral nucleus of amygdala is found to play a role in processing information that leads to the phobia and even anxiety disorder (File, S., Gonzalez, L. & Gallant, R., 1998). Meanwhile, studies have shown that the lateral nucleus of the amygdala is essential in auditory fear conditioning as neurons in the amygdala lateral respond to auditory stimuli (Romanski, L. M., Clugnet, M.-C., Bordi, F., & LeDoux, J. E., 1993). Therefore, we would like to discuss if there is any difference in the gene expression in these two regions of amygdala so that they function differently. 

Projection neurons are mostly excitatory neruons in brain regions. Projection neurons’ axons extend from the neuronal cell body within the central nervous system to other distant regions of the CNS (Nusbaum M.P., 2009). They are involved in various projection pathways which would lead to different functions in their target structure. Based on the properties of projection neurons, we believe that learning more about electrophysiology features in each region may help us better understand the different function of basolateral and lateral amygdala. Moreover, pyramidal cells are projection neuron as principal neurons in the basolateral nucleus of amygdala to generate emotional responses based on the ratio of excitatory and inhibitory inputs to the cells (Muller, J.F., Mascagni, F. and McDonald, A.J., 2006). Therefore, we decided to choose the pyrmidal cells in basolateral amygdala and compare its electrophysiological features with projection neurons in lateral amygdala to reflect the function of the region. 

In order to analyze the differences in gene expression and electrophysiology, we will use the Allen ISH and NeuroElectro datasets. Allen ISH is part of the Allen Brain Atlas which is a publicly published online resource containing various information about the human and mouse brain. The in situ hybridization image data with cellular-level detail is gained from the adult mouse brain to show the genome-wide atlas of its gene expression. Differential search demonstrates relative gene expression in one regions of the brain comparing to the other region. ### Gene expression dataframe

The Allen Brain Atlas API is used to pull differential gene expression from the mouse in situ hybridization (ISH) data. The differential expression will tell, relatively, how strong these differences are between the target structure and the contrast structure by the value called 'fold change'. Here, the first set of data uses basolateral amygdalar nucleus as the target structure, and the lateral amygdalar nucleus as contrast structure. The second set of data uses lateral amygdalar nucleus as the target structure, and the basolateral amygdalar nucleus as contrast structure. Both set of data pulled from the API are then converted into a Pandas Dataframe so that it can more easily be viewed and analyzed later.

```python
# Code to import the needed toolboxes from the Allensdk
from allensdk.core.cell_types_cache import CellTypesCache
from allensdk.api.queries.cell_types_api import CellTypesApi
```

```python
import requests
import json
```

Basolateral amygdalar nucleus vs Lateral amygdalar nucleus

```python
# Set up API to retrieve mouse ISH data
service = 'http://api.brain-map.org/api/v2/data/query.json?criteria=' 
structures_1 = ['Basolateral amygdalar nucleus']
structures_2 = ['Lateral amygdalar nucleus']
ontology = 'Mouse Brain Atlas'

# Get IDs for examined structures
ids_1 = []
ids_2 = []
for item in structures_1:
    structure_id = requests.get('%smodel::Structure,\
        rma::criteria,[name$il\'%s\'],ontology[name$eq\'%s\']' % (service,item,ontology)).json()['msg'][0]['id']
    ids_1.append(str(structure_id))
for item in structures_2:
    structure_id = requests.get('%smodel::Structure,\
        rma::criteria,[name$il\'%s\'],ontology[name$eq\'%s\']' % (service,item,ontology)).json()['msg'][0]['id']
    ids_2.append(str(structure_id))
# Create a single string for each set of structure IDs to use in API
temp1 = ''
temp2 = ''
for i in ids_1:
    temp1 += i + ','
for i in ids_2:
    temp2 += i + ','
structures1 = temp1[0:-1]
structures2 = temp2[0:-1]

# Get results for differential search and put into a pandas dataframe
all_results = []
test = True
start_row = 0
# API is only able to retrieve 2000 rows at once, loop until none remain
while test:
    result = requests.get('%sservice::mouse_differential[set$eq\'mouse\'][structures1$eq\'%s\'][structures2$eq\'%s\'][start_row$eq\'%s\']' \
                          % (service, structures1, structures2, str(start_row))).json()
    if result['success'] == True:
        all_results += result['msg']
        start_row += 2000
    else:
        test = False
mouse_differential = pd.DataFrame(all_results)

mouse_differential
```

Lateral amygdalar nucleus vs. Basolateral amygdalar nucleus

```python
temp1_LA = ''
temp2_BLA = ''
for i in ids_2:
    temp1_LA += i + ','
for i in ids_1:
    temp2_BLA += i + ','
structures1_LA = temp1_LA[0:-1]
structures2_BLA = temp2_BLA[0:-1]

# Get results for differential search and put into a pandas dataframe
all_results = []
test = True
start_row = 0
# API is only able to retrieve 2000 rows at once, loop until none remain
while test:
    result = requests.get('%sservice::mouse_differential[set$eq\'mouse\'][structures1$eq\'%s\'][structures2$eq\'%s\'][start_row$eq\'%s\']' \
                          % (service, structures1_LA, structures2_BLA, str(start_row))).json()
    if result['success'] == True:
        all_results += result['msg']
        start_row += 2000
    else:
        test = False
mouse_differential_LA_BLA = pd.DataFrame(all_results)

mouse_differential_LA_BLA
```

We chose to plot swarmplots to better visualize individual data points for electrphysiological features. However, by looking into the current data, we might still have some conclusions about certain electrophysiologicla features. For both basolateral and lateral amygdala projection neuron, their resting membrane potential seems to be lower than -60mV, and their spike threshold might be above -45mV. It seems that the AHP amplitude for projection neurons in both regions might be lower than 20. However, there's an outlier for basolateral data which is at about 34. We can't exclude the possibility that in fact, most projection neuron in basolateral amygdala would have a AHP amplitude at about 34 and only a small amount would be lower than 20. The AP half width of basolateral projection neuron seems to vary a lot with an average about 1.2, while the average of lateral projection neuron is about 1.4. The input resistance for two type of projection neuron seems to differ as well, with that basolateral projection neuron usually lower than 200 but lateral projection neurons here have an average about 200. Based on the data we have, it seems that there are electrophysiological differences of projection neuron in the basolateral and lateral amygdala, which can help explain the function difference of the two regions.   

As mentioned early, the difference in Kv channel interacting protein 3, calsenilin between these two regions may responsible for difference in electrophysiology. Calsenilin, as a voltage-voted potassium channel, would provide the outward currents of action potentials when the action potential reaches the peak, and this is essential in determining the electrical excitability of neurons. The voltage-voted potassium channel would open from peak potential through hyperpolarization to enable potassium to move from the intracellular fluid to the extracellular fluid. With different amount of calsenilin, the rate and duration of the action potential may vary as well, which may lead to the difference in the electrophysiology rate.

# Background and Prior Work

Alzheimer's disease (AD) is a neurodegenerative disease that progressively
worsens over a period of time. By exploring the effects of
differential gene expression in both cell types and brain regions, it would help provide a greater
understanding of how certain gene expressions play a role in the progression of
Alzheimer's and how it might be linked to cognitive decline in Alzheimer’s patients. It
would help us gain insights and conclude which brain regions and cell types show differential expression of genes in Alzheimer’s Disease. 

Studies have found that there are several genes associated with the development of Alzheimer’s disease. The early-onset of AD primarily involves mutations in three genes —  amyloid-β precursor protein (APP), presenilin 1 (PSEN1), and presenilin 2 (PSEN2) — while the apolipoprotein E (APOE) gene is considered a risk factor for late-onset AD (Parasuraman et al. 2002). The APOE gene, which is responsible for coding a protein that helps regulate and transport cholesterol through the bloodstream, can be inherited as any of three major types in the form of e2, e3, or e4 alleles (Parasuraman et al. 2002). Among these types, the e4 allele of the APOE gene is the greatest factor associated with increasing the chance of receiving late-onset AD, but while it is a risk gene, it is not a definitive biomarker of the disease. Also, there have been studies done with magnetic resonance imaging that shows that Alzheimer's patients have a significantly smaller hippocampus relative to that of a healthy person (Convit et al., 1997; Kesslak, Nacclioglu, & Cotman, 1991; Killiany, Moss, Albert, & Tamas, 1993). Furthermore, studies conducted with twins who have the APOE genotype reveals that those who have the e4 allele (n=6) have structurally smaller hippocampus than those who do not have the e4 allele (n=14). This conclusion suggests that there is some level of differential expression of the APOE-e4 gene that affects the structure, function, and volume of the hippocampus. Thus, it gives us some insight that some brain regions and cell types may indeed have differential gene expression as a result of Alzheimer's Disease. 

Previous studies have also discovered that the APOE-e4 gene affects certain cell types in the brain. It has been suggested that the varying expression of the APOE-e4 gene in neural and glial cells can lead to severe AD (Lin et. al, 2018). As a consequence of the varied gene expression of APOE-e4, microglial cells activate inflammatory genes and have less efficient clearing of Aβ42 in AD brains. Additional research has found that the four major types of cells that are typically affected and could be related to Alzheimer’s disease are through the pyramidal neuron, astrocyte, oligodendrocyte and microglia cell types (Miller et. al, 2013). Other studies have shown that glial cells are involved in glutamatergic transmission and that GABAergic cell content rises as age increases, so this may also indicate a relationship involving expression of glutamatergic and GABAergic cell types with AD-associated genes (Martínez-Lozada, et. al, 2015; Olga, et. al, 2019). We would like to investigate the differential gene expression levels of the APP, PSEN1, PSEN2, and APOE genes across the major cell types and brain regions in Alzheimer’s Disease. 

The first dataset that we used is the Allen Brain Atlas microarray data to see how APP, PSEN1, PSEN2, and APOE gene expressions varied in different brain regions. There were six different donors, varying from age, gender, and ethnicity, that contributed to the microarray data collection. The dataset recorded 27 unique top level brain regions analyzed and within those 27, there were 108 unique specific brain structures that the expression of genes were measured for. In total, there were 1014 observations of expression for each gene. The dataset for each gene provided information about their expression values, the metadata for each column in the expression data, and the metadata for the probes referenced. The expression data contained the expression values with Z-score normalization for each gene probe. The column metadata supplied the donor’s name, age, and ID, the expression MRI coordinates, the brain structure’s name, ID, and color of expression. The probe metadata contained information about the gene probe’s name and ID, the gene’s name, symbol, and ID, and lastly the chromosome number and position.

The second dataset that we will be using is the RNA Sequencing of Cell Types in order to show  differential expression of APP, PSEN1, PSEN2 and APOE genes in different cell types. The nuclei that were used for sampling were from postmortem and neurosurgical (MTG only) donors. The dataset has 76,533 total nuclei to see the diversity of cell types in the primary motor cortex (M1). Also, in this dataset, there were 127 transcriptomic cell types that they were able to identify. In this dataset, we chose to look at the metadata and trimmed means. Within the metadata, it includes information about the cell types and their order, label, and color column. The trimmed means includes the trimmed means of gene expression per cell type.

## Conclusion & Discussion

Based upon the 20 scatter plots of expression level of the 4 genes vs. age of the donors in 5 different brain regions; and their corresponding p-values for correlation, we could conclude that there is a positive correlation between the expression level of the PSEN1 gene and age of the donor in the frontal and temporal lobes since the corresponding p-values are 0.033 and 0.035 and r-value respectively for these two dataset and their r-values are 0.848 and 0.951 respectively .  For the remaining 18 datasets, all their p-values are greater than 0.05, thus we could not conclude that there is a significant correlation within these datasets. 

Based on our boxplot data of the APP gene in the first dataset, the cerebellum cortex tends to have the lowest expression value in contrast to the claustrum region having the highest expression value. For the boxplots of PSEN1, hippocampal formation had the lowest expression value while white matter had the highest. For PSEN2, white matter was the lowest while the dorsal thalamus had the highest expression value. For APOE, the globus pallidus had the highest value and sulci & spaces, the lowest. The heatmap of these gene expression values based on the brain regions also provide additional visualization of each gene’s highest and lowest levels of expression to support this.  These results indicate a level of significance to which there are differential expressions based on the regions of the brain for genes associated with Alzeimer’s  disease.

According to the boxplot and heatmap for the second dataset, it clearly shows that there is a significant level of gene expression across the APP gene and PSEN 1 gene. It shows that the GABAergic and Glutamatergic neurons of these cell types within the primary motor cortex showed the highest level of expression. Also,  it shows that PSEN1 and APOE genes did not have any expression in the cell types of the primary motor cortex, aside from a few outliers present in the APOE gene. Throughout the dataset, it showed that there is indeed differential expression of genes within the primary motor cortex. This can be seen with the high and low expression levels between the APP, PSEN1, PSEN2, and APOE gene. 

There were a couple of limitations to our analysis with the first being that our data lacked information from people with Alzheimer’s disease. Furthermore, the sample size is very small as there were only six donors for the entire dataset on the gene expression levels in brain regions. Another major limitation is that the cell type dataset was limited to one cortex of the brain instead of multiple cortexes. It is also notable that the other factors besides the age of the donors are not controlled in this data set (e.g: gender, ethnicity, etc.).

Additional experiments that we could do would be to examine these four gene expressions within patients that have AD as well as gathering additional data on the cell types associated with the brain regions. Increasing the sample size across a broader range of age groups would also be another way we can obtain a stronger correlation. Other experiments that would be interesting would be to see if gender or ethnicity creates any disparity in the amount of risk of AD onset by analyzing any difference in gene expression in the brain and for cell types. We could also investigate the correlation between the age and particular cell types in the brain if we have a dataset of the gene expression in various cell types labelled with the donors’ ages.
